```python
import numpy
import matplotlib.pyplot as plot

x=numpy.arrange(0,10,0.1)
sinx=numpy.sin(x)
plot.plot(x,sin(x))
h=0.00001
fdx=(numpy.sin(x+h)-numpy.sin(x))/h
plot.plot(x,fdx)
plot.show()
```
